```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Classe gráfico
class Grafico:
  def __init__(self, listade_despesas):
    self.listade_despesas = listade_despesas
    self.imprimir_graficos()

  def padrao_do_grafico(self):
    plt.xlabel('Dia')
    plt.ylabel('Gasto (R$)')
    plt.title('Gráfico dos gastos')

  def imprimir_graficos(self):
    self.padrao_do_grafico()
    for gast in self.listade_despesas:
      mLista = gast.dicionario.items()
      cor = gast.cor
      nome = gast.nome
      x, y = zip(*mLista)
      plt.plot(x, y, label = nome, marker='o',
               markerfacecolor = 'blue',
               markersize = 12,
               color = cor,
               linewidth = 4)
    plt.legend()
    plt.show()

  def regressao_linear(self, id_grafico):
    gast = self.listade_despesas[id_grafico]
    mLista = gast.dicionario.items()
    cor = gast.cor
    nome = gast.nome
    dia, valor = zip(*mLista)
    dia = np.array(dia)
    valor = np.array(valor)
    dia = dia.reshape(-1, 1)
    valor = valor.reshape(-1, 1)
    regressao = LinearRegression()
    regressao.fit(X = dia, y = valor)
    plt.plot(dia, regressao.predict(dia),
             color='blue',
             label = "Regressão Linear")

    x, y = zip(*mLista)
    plt.plot(x, y, label = nome+str(" - original"),
             marker = 'o',
             markerfacecolor = 'olive',
             markersize = 12,
             color = cor,
             linewidth = 4)

    plt.legend()
    plt.show()
```

```python
# Classe dicionário
class Dicionario:

    def __init__(self, dicionario, cor, nome):
        self.dicionario = dicionario
        self.cor = cor
        self.nome = nome
```

```python
# Despesas + Gráfico + Regressão linear


despesa_comida = Dicionario({1:70,2:50,3:100,4:200,5:300},'skyblue','Comida')


despesa_transporte = Dicionario({1:70,2:95,3:115,4:200,5:450},'red','Transporte')


despesa_roupa = Dicionario({1:100,2:60,3:55,4:100,5:250},'olive','Roupas')


listade_despesas = [despesa_comida, despesa_transporte, despesa_roupa]


grafico = Grafico(listade_despesas)




id_despesas = 0
grafico.regressao_linear(id_despesas)
```
